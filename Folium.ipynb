```python
import folium
import json
import numpy as np
import pandas as pd
import geopandas as gpd
import json
```

```python
geo_json_data = json.load(open('punjab_districts.geojson'))
geo_json_data['features'][2]
```

```python
df_gp = gpd.read_file('punjab_districts.geojson')
df_gp = df_gp[['district','geometry']]
df_gp.drop(df_gp.index[21] , inplace=True)
```

```python
df_gp.reset_index(drop=True,inplace=True)
df_gp.head(4)
```

```python
df = pd.read_csv('data2.csv')
df.head(4)
```

```python
# create a numpy array of length 6 and has linear spacing from the minium total immigration to the maximum total immigration
threshold_scale = np.linspace(df['Fire'].min(),
                              df['Fire'].max(),
                              6, dtype=int)
threshold_scale = threshold_scale.tolist() # change the numpy array to a list

#Create a map:
Map = folium.Map(location=[31.17, 72.7097], zoom_start=6.5, tiles='Stamen toner')

#Create a layer
Map.choropleth(geo_data="punjab_districts.geojson",
                     fill_color='YlGn', fill_opacity=2, line_opacity=0.2,
#                      threshold_scale = [60000,70000,80000,90000,100000,110000],
                     threshold_scale = threshold_scale,
                     data = df,
                     key_on='feature.properties.district',
                     columns = ['district', 'Fire'],
                     legend_name='Fire Incidents'
                     
                     )

for i in range(len(df_gp)):
    gs = folium.GeoJson(df_gp.iloc[i:i+1],
                        style_function=lambda i: {
#                         'fillColor': 'black' if 'a' in i['properties']['district'].lower() else '#ffff00',
#                         'color' : 'black',
#                         'weight' : 1.5,
                        })
    label = '{}: {} Fire incidents '.format(df['district'][i], df['Fire'][i])
    folium.Popup(label).add_to(gs)
    gs.add_to(Map)
Map
```

```python
Map.save(outfile='Map.html')
```
